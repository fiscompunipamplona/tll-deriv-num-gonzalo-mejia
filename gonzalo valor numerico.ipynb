partiendo de la segunda ley de Newton:

\begin{equation}
\ F=F_{g} + F_{a}
\end{equation}

encontramos la ecuacion diferencial 

\begin{equation}
\frac{dv}{dt}= g- \frac{c}{m}v
\end{equation}

con la ecuacion diferencial hallamos la solucion analitica

\begin{equation}
\ v= \frac{gm}{c}\left ( 1- e^{\frac{-c}{m}t} \right )
\end{equation}
    

```python
from pylab import plot, show, legend
from math import exp, fabs
```

```python
vnum=[]
tp=[]
vana=[]

dlt=1
g=9.8
m=68.1
c=12.5
vi=0

for t in range (21):
    
    vi=vi+((c/m)*vi-g)*dlt
    v=(m*g/c)*(1-exp(c*t/m))
    
    vnum.append (vi)
    tp.append (t)
    vana.append (v)
    

plot (tp,vana, label=' valor analitico')
plot (tp,vnum, label=' valor numerico')
legend(loc='lower left')
show
```

```python
Anum=[]
Anum2=[]
Anum3=[]
time1=[]
Aana=[]



for n in range (1,20):
    
    a= -g*exp((c*n)/m) # valor analitica
    v=(vnum[n]-vnum[n-1])/dlt #valor numerica hacia  atras
    
    time1.append (n)
    Anum.append (v)
    Aana.append (a)
    
for n in range (1,20):

    v=(vnum[n+1]-vnum[n])/dlt #valor numerica hacia  adelante
  
    Anum2.append (v)

for n in range (1,20):

    v=(vnum[n+1]-vnum[n-1])/2*dlt #valor numerica centrada
  
    Anum3.append (v)

    
plot (time1,Aana, label=' valor analitico')
plot (time1,Anum, label=' valor numerico atras')
plot (time1,Anum2, label='valor numerico adelante')
plot (time1,Anum3, label='valor numerico centrado')
legend(loc='lower left')
show
```
